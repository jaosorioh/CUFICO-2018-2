```
import ROOT
```

```
from array import array
```

```
MyTree = ROOT.TTree( 'tree1', 'tree para tres variables' )
```

```
x = array( 'f', [ 0 ] )
y = array( 'f', [ 0 ] )
z = array( 'f', [ 0 ] )
```

```
MyTree.Branch( 'x', x, 'x/F' )
MyTree.Branch( 'y', y, 'y/F' )
MyTree.Branch( 'z', z, 'z/F' )
```

```
import numpy as np
```

```
for i in range(5000):
  x[0] = (np.random.randn() + 1)*5
  y[0] = (np.random.randn() + 1)*5
  z[0] = np.random.rand()*10
  MyTree.Fill()

print(z)
```

```
#Grafica: x*y
c1 = ROOT.TCanvas( 'c1', 'My First Data example', 200, 10, 700, 900 )
c1.SetFillColor( 18 )
MyTree.Draw("x*y>>h1(20, 0, 100)")
c1.Update()
from ROOT import gROOT 
gROOT.GetListOfCanvases().Draw()
```

```
#Grafica: x*z
c1 = ROOT.TCanvas( 'c1', 'My First Data example', 200, 10, 700, 900 )
c1.SetFillColor( 18 )
MyTree.Draw("x*z>>h1(20, 0, 100)")
c1.Update()
from ROOT import gROOT 
gROOT.GetListOfCanvases().Draw()
```

```
#Grafica: x*y*z
c1 = ROOT.TCanvas( 'c1', 'My First Data example', 200, 10, 700, 900 )
c1.SetFillColor( 18 )
MyTree.Draw("x*y*z>>h1(100, 0, 1000)")
c1.Update()
from ROOT import gROOT 
gROOT.GetListOfCanvases().Draw()
```

```
#Grafica: x cercano a 5 
c1 = ROOT.TCanvas( 'c1', 'My First Data example', 200, 10, 700, 900 )
c1.SetFillColor( 18 )
MyTree.Draw("x>>h1(20, 0, 10)", "abs(z - 5)<=0.5")
c1.Update()
from ROOT import gROOT 
gROOT.GetListOfCanvases().Draw()
```

```
#Grafica: x:z para y cercano a 5
c1 = ROOT.TCanvas( 'c1', 'My First Data example', 200, 10, 700, 900 )
c1.SetFillColor( 18 )
MyTree.Draw("x:z>>h1(20, 0, 10, 20, 0, 10)", "abs(y-5)<=0.5", "lego")
c1.Update()
from ROOT import gROOT 
gROOT.GetListOfCanvases().Draw()
```

```
#Grafica: x:y:z
c1 = ROOT.TCanvas( 'c1', 'My First Data example', 200, 10, 700, 900 )
c1.SetFillColor( 18 )
MyTree.Draw("x:y:z>>h1(10, 0, 10, 10, 0, 10, 10, 0, 10)", "", "lego")
c1.Update()
from ROOT import gROOT 
gROOT.GetListOfCanvases().Draw()
```
